```
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize
import time
```

```
api_key = "ZI0iUaNUAZvGVZt7ufAfvEGS2FBlaF2V"
```

```
response = requests.get("https://app.ticketmaster.com/discovery/v2/classifications.json?apikey=" + api_key)
data_test = response.json()
```

```
#shows the different classification names (sports, concerts, etc.)
data_test
```

```
#get a dictionary of all classifications and their ids
class_dict = {"segments": []}

for classification in data_test["_embedded"]["classifications"]:  
  if ("segment" in classification): 
    class_dict["segments"].append({"name" : classification["segment"]["name"], 
                                "id": classification["segment"]["id"]})

class_dict
```

```
#not interested in undefined
del class_dict["segments"][4]
```

```
import time 

#get a dictionary of subgenres associated with segments 
for segment in class_dict["segments"]: 
  genres = []
  response = requests.get("https://app.ticketmaster.com/discovery/v2/classifications/segments/%s.json?apikey=%s" % (segment["id"], api_key))
  resp = response.json()
  for genre in resp["_embedded"]["genres"]: 
    genres.append({"name" : genre["name"], 
                   "id" : genre["id"]})  
  segment.update({ "genres" : genres})
  time.sleep(0.5)

class_dict
```

```
#focus on music genre
music = class_dict["segments"][2]
music
```

```
#getting music by genre
m_resp = requests.get("https://app.ticketmaster.com/discovery/v2/events.json?classificationName=music&genreId=KnvZfZ7vAv1&countryCode=US&size=200&apikey=" + api_key)
music_json = m_resp.json()
df_music = json_normalize(music_json["_embedded"]["events"])
```

```
#manipulating/cleaning up smaller df for training 
df_music_train = df_music[df_music["priceRanges"].isna() == False].copy()

#add more helpful columns 
df_music_train["priceMin"] = df_music_train["priceRanges"].map(lambda x: x[0]['min'] if 'min' in x[0].keys() else -9999)
df_music_train["priceMax"] = df_music_train["priceRanges"].map(lambda x: x[0]["max"] if 'max' in x[0].keys() else -9999)
df_music_train["subGenre"] = df_music_train["classifications"].map(lambda x: x[0]["subGenre"]["name"] if 'subGenre' in x[0].keys() else "None")
df_music_train["city"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["city"]["name"])
df_music_train["state"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["state"]["name"])
df_music_train["venueName"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["name"])
df_music_train["latitude"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["location"]["latitude"])
df_music_train["longitude"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["location"]["longitude"])
df_music_train["venueUpcoming"] = df_music_train["_embedded.venues"].map(lambda x: x[0]["upcomingEvents"]["_total"])
df_music_train["attractionName"] = df_music_train["_embedded.attractions"].map(lambda x: x[0]["name"])

artists_full = []
num_artists = []

df_music_train.reset_index(inplace=True)

for i in range(len(df_music_train)):
  artists = []
  events = df_music_train["_embedded.attractions"][i]

  for artisit in events:
    artists.append(artisit["name"])

  artists_full.append(artists)
  num_artists.append(len(artists))
  
df_music_train["artists"] = artists_full
df_music_train["num.artists"] = num_artists


#drop unnecessary columns 
df_music_train = df_music_train.drop(["test", "url", "images", "_links.self.href", "_links.attractions", "_links.venues", "seatmap.staticUrl"], axis=1)
df_music_good = df_music_train[["name", "type", "promoter.name", "info", "pleaseNote", 
                                "priceMin", "priceMax", "subGenre", "city", "state", "venueName", "attractionName", "artists", "num.artists", 
                                "latitude", "longitude", "venueUpcoming"]].copy()

df_music_good.drop(df_music_good[df_music_good["priceMin"] == -9999].index, inplace=True)
df_music_good.drop(df_music_good[df_music_good["priceMax"] == 0].index, inplace=True)
df_music_good["meanPrice"] = (df_music_good["priceMin"] + df_music_good["priceMax"])/2

df_music_good["date"] = pd.to_datetime(df_music_train["dates.start.localDate"], format="%Y-%m-%d")
```
